```python
#Задание 1

#Для датафрейма log из материалов занятия создайте столбец source_type по следующим правилам:

#если источник traffic_source равен yandex или google, то в source_type ставится organic
#для источников paid и email из России - ставим ad
#для источников paid и email не из России - ставим other
#все остальные варианты берем из traffic_source без изменений
```

```python
log=pd.read_csv('visit_log.csv', sep=';')
log.head()
```

```python
log.loc[(log.traffic_source == 'yandex') | (log.traffic_source == 'google'), 'source_type'] = 'organic'
log.loc[((log.traffic_source == 'paid') | (log.traffic_source == 'email')) & (log.region=='Russia'), 'source_type'] = 'ad'
log.loc[((log.traffic_source == 'paid') | (log.traffic_source == 'email')) & (log.region!='Russia'), 'source_type'] = 'other'
```

```python
log.loc[log.source_type.isnull(), 'source_type']=log['traffic_source']
```

```python
#Задание 3

#Используйте файл с оценками фильмов ml-latest-small/ratings.csv. 
#Посчитайте среднее время жизни пользователей, которые выставили более 100 оценок. 
#Под временем жизни понимается разница между максимальным и минимальным значением столбца timestamp для данного значения userId.
```

```python
ratings=pd.read_csv('ml-latest-small/ratings.csv')
ratings.head()
```

```python
ratings_count=ratings.groupby('userId').count().reset_index()[['userId','rating']]
ratings_count.head()
```

```python
#юзеры с более 100 оценок
count_more_than_100=ratings_count[ratings_count['rating']>100]
count_more_than_100.head(10)
```

```python
#max время жизни
rating_max=ratings.groupby('userId').max().reset_index()[['userId','timestamp']]
rating_max.rename(columns={'timestamp': 'timestamp_max'}, inplace=True)
rating_max.head()
```

```python
#min время жизни
rating_min=ratings.groupby('userId').min().reset_index()[['userId','timestamp']]
rating_min.rename(columns={'timestamp': 'timestamp_min'}, inplace=True)
rating_min.head()
```

```python
avg_timestamp=count_more_than_100.merge(rating_max, how='inner').merge(rating_min, how='inner')
```

```python
#разница между максимальным и минимальным
def average(row):
    mean_=(row['timestamp_max']-row['timestamp_min'])
    return mean_
```

```python
avg_timestamp['avg']=avg_timestamp.apply(average, axis=1)
```

```python
rzd = pd.DataFrame(
    {
        'client_id': [111, 112, 113, 114, 115],
        'rzd_revenue': [1093, 2810, 10283, 5774, 981]
    }
)
rzd
```

```python
air = pd.DataFrame(
    {
        'client_id': [115, 116, 117, 118],
        'air_revenue': [81, 4, 13, 173]
    }
)
air
```
